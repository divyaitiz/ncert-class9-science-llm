# Text Chunking for 'Think and Act' Sections

This script is designed to process a text file (`thinkandact.txt`) and extract specific sections, which are identified by the "thinkandact:" prefix. It then structures these extracted chunks into a JSON format, adding chapter and page number metadata.

```python
import json
import re

# ---- CONFIG ----
INPUT_FILE = "thinkandact.txt"
OUTPUT_FILE = "chapter7_72_thinkandact_output.json"

CHAPTER = "7"   # manually set
PAGE_NO = "72"  # manually set
# ----------------

def parse_file(filepath):
    entries = []
    current_content = []

    with open(filepath, "r", encoding="utf-8") as f:
        for line in f:
            stripped = line.strip()

            if stripped.lower().startswith("thinkandact:"):
                # save previous block if exists
                if current_content:
                    entries.append(" ".join(current_content).strip())
                    current_content = []

                # start new block (remove label)
                content = stripped[len("thinkandact:"):].strip()
                if content:
                    current_content.append(content)
            else:
                current_content.append(stripped)

        # last block
        if current_content:
            entries.append(" ".join(current_content).strip())

    return entries


def build_json(entries):
    structured = []
    for content in entries:
        # Remove 'Think and Act' from the beginning of the content, case-insensitively
        cleaned_content = re.sub(r'^\s*Think and Act\s*', '', content, flags=re.IGNORECASE).strip()
        structured.append({
            "chapter": CHAPTER,
            "pg no": PAGE_NO,
            "thinkandact": cleaned_content
        })
    return structured


if __name__ == "__main__":
    entries = parse_file(INPUT_FILE)
    result = build_json(entries)

    with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
        json.dump(result, f, indent=2, ensure_ascii=False)

    print(f"Processed {len(result)} entries → {OUTPUT_FILE}")
```

### **Configuration**

The following variables are set to configure the input and output files, as well as metadata (chapter and page number) to be included in the output JSON. These are currently manually set.

### **`parse_file(filepath)` Function**

This function reads the specified text file line by line. It identifies "thinkandact:" entries, treating them as separators or indicators for new content blocks. It concatenates the lines between these markers (or from the start/end of the file) into distinct text entries and returns them as a list.

### **`build_json(entries)` Function**

This function takes the list of text entries generated by `parse_file`. For each entry, it constructs a dictionary containing the predefined `CHAPTER`, `PAGE_NO`, and the extracted `thinkandact` content. It returns a list of these structured dictionaries, ready for JSON serialization.

### **Main Execution Block**

When the script is run, this block orchestrates the workflow:
1. Calls `parse_file` to extract raw entries from `INPUT_FILE`.
2. Calls `build_json` to structure these entries with metadata.
3. Writes the final structured data to `OUTPUT_FILE` in a human-readable JSON format.
