# Activity Parsing Function

This section defines the initial Python function `chunk_activities_to_json` responsible for parsing the `activities.txt` file. It identifies individual activities and attempts to extract page numbers and activity text into a structured dictionary format.

```python
import re

def chunk_activities_to_json(text):
    lines = text.splitlines()

    activity_pattern = re.compile(r'^Activity\s+(\d+\.\d+)', re.I)
    page_pattern = re.compile(r'^\d{2,3}$')  # page numbers like 73, 74

    chunks = []
    current_chunk_lines = []
    current_activity_number = None
    current_page = None

    for line in lines:
        line = line.strip()

        if not line:
            continue

        # Detect page number
        if page_pattern.match(line):
            current_page = int(line)
            continue

        # Detect new activity
        activity_match = activity_pattern.match(line)
        if activity_match:
            # Save previous chunk
            if current_chunk_lines and current_activity_number:
                chunks.append({
                    "type": "activity",
                    "activity_number": current_activity_number,
                    "text": "\n".join(current_chunk_lines),
                    "page": current_page
                })
                current_chunk_lines = []

            # Start new activity
            current_activity_number = activity_match.group(1)
            current_chunk_lines.append(line)

        else:
            current_chunk_lines.append(line)

    # Save last chunk
    if current_chunk_lines and current_activity_number:
        chunks.append({
            "type": "activity",
            "activity_number": current_activity_number,
            "text": "\n".join(current_chunk_lines),
            "page": current_page
        })

    return chunks
```

## Initial Activity Extraction and Page Detection

The `chunk_activities_to_json` function is called to process the `activities.txt` file. The output below shows the initially extracted page numbers for each activity. It was observed that these automatically extracted page numbers were often incorrect.

```python
with open("activities.txt", "r") as f:
    text = f.read()

chunks = chunk_activities_to_json(text)

for c in chunks:
    print(c['page'])
```
