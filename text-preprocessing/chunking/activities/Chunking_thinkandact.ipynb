# Text Chunking for 'Think and Act' Sections

```python
def chunk_think_and_act(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        lines = [line.strip() for line in f.readlines() if line.strip()]

    chunks = []
    i = 0

    while i < len(lines):
        # Detect section header
        if lines[i].lower() == "think and act":
            chunks.append({
                "type": "header",
                "content": lines[i]
            })
            i += 1

            # Collect paragraph until next header-like line or end
            paragraph_lines = []
            while i < len(lines) and lines[i].lower() != "think and act":
                paragraph_lines.append(lines[i])
                i += 1

            if paragraph_lines:
                chunks.append({
                    "type": "paragraph",
                    "content": " ".join(paragraph_lines)
                })
        else:
            i += 1

    return chunks


# Example usage
chunks = chunk_think_and_act("/content/thinkandact.txt")

for c in chunks:
    print(c["type"], ":", c["content"])
```

### Function: `chunk_think_and_act`

This Python function `chunk_think_and_act` is designed to process a text file and extract specific sections. It looks for lines containing 'Think and Act' (case-insensitive) as headers and then collects the subsequent paragraph lines until another 'Think and Act' header or the end of the file.

The function returns a list of dictionaries, where each dictionary represents a chunk. A chunk can be of type 'header' (for 'Think and Act' lines) or 'paragraph' (for the text following a header).
